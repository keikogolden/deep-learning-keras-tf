Since the real world doesn't work like linear expression, the activation fucntions like Sigmoid, tanh, ReLu, LeakyRelu help you to build such a non liner expression.

<h3 style='color:blue' align='center'>Implementation of activation functions in python</h3>

<h3 style='color:purple'>Sigmoid</h3>

Use Sigmoid in output layer. All other places try to use Tanh.

```python
import math

def sigmoid(x):
    return 1 / (1 + math.exp(-x))
```

```python
sigmoid(100)
```

```python
sigmoid(200)
```

```python
sigmoid(1)
```

```python
sigmoid(-56)
```

```python
sigmoid(-300)
```

```python
sigmoid(0.5)
```

```python
sigmoid(0.05)
```

<h3 style='color:purple'>tanh</h3>

Tanh: While Sigmoid give the output between 0 to 1, tanh ives the output between -1 and 1.

Use Sigmoid in output layer. All other places try to use Tanh.Because Tenh will calculate a mean of 0 and it will center your data.

```python
def tanh(x):
    return (math.exp(x) - math.exp(-x)) / (math.exp(x) + math.exp(-x))
```

```python
tanh(-56)
```

```python
tanh(-5)
```

```python
tanh(-70)
```

```python
tanh(50)
```

```python
tanh(100)
```

```python
tanh(1)
```

#### Issues with Sigmoid and Tanh: Vanishing Gradients
Because of the Vanishing Gradients, it makes the learning process very slow. That's why they came up to the function called ReLU.


<h3 style='color:purple'>ReLU</h3>

If your value is less than zero, the output (y value) is zero.
If your value is more than zero, the output (y value) is same as the value.
If I put 2, I get 2 as an output.
If I put -2, I get 0 as an output.


##### For hidden layers if you are not sure which activation function to use, just use ReLu as your default choice.

```python
def relu(x):
    return max(0,x)
```

```python
relu(-100)
```

```python
relu(8)
```

However, ReLu also has Vanishing Gradients problem because where if value is less than zero, the derivative again is zero, like Sigmooid and Tanh.For that leason, there is another activation calles Leaky ReLU.

<h3 style='color:purple'>Leaky ReLU</h3>

Leaky ReLu is where you know there is still some sort of linear line, it is still trying to reduce the value close to zero, but the equation not instead of 0 is 0.1.


```python
def leaky_relu(x):
    return max(0.1*x,x)
```

```python
leaky_relu(-100)
```

```python
leaky_relu(8)
```
