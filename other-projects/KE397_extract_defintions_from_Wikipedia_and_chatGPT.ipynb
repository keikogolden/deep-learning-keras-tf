```python
import pandas as pd
import requests
```

```python
def get_definitions_from_wikipedia(skill:str):
    
    skill_underscored = "_".join(skill.split(" "))
    url = f'''https://en.wikipedia.org/wiki/{skill_underscored}'''
    res = requests.get(url)
    return res
```

```python
path = "C:/Users/KeikoGolden/ComputationalLiguistics/KE397_1500skillsCandidate/anay.xlsx"
input_df = pd.read_excel(path).dropna(how='all')
input_df
```

```python
# Depending on where the skill title is in input_df, change the column name in the input_df.
from bs4 import BeautifulSoup
import pandas as pd

results_df = pd.DataFrame()

for title in input_df['skill_title']:
    response = get_definitions_from_wikipedia(title)
    response_text = response.text
    soup = BeautifulSoup(response_text)
    #paras = soup.find_all(["h1", "p"])
    paras = soup.find_all(["p"])
    texts = [p.text for p in paras]
    wiki_def_df = pd.DataFrame({'wiki_def':texts[:1]})
    results_df = pd.concat([results_df, wiki_def_df])
```
