```python
import spacy
import pandas as pd
```

```python
df = pd.read_excel("C:/Users/KeikoGolden/ComputationalLiguistics/KE413/proedge_delivery.xlsx")
df
```

```python
# Extract unigramsfro the input file
nlp = spacy.load("en_core_web_lg")

unigrams = []
for skill in df.skill_title:
    doc = nlp(skill)
    if len(doc)== 1:                               
        unigrams.append(doc)
    
```

```python
result_df = pd.DataFrame({'unigrams': unigrams})
result_df.to_excel('unigrams.xlsx')
```
