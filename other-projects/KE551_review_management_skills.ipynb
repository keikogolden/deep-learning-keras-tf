### 2-1 Find similar term to 'management' from resume corpus and from non-skyhive db, then create labels

```python
def findSimilarity_noDup_withScore(modifier, pos, similarity_rate):

    
    nlp = spacy.load('en_core_web_lg')
    
    excellent_doc = nlp(modifier)
    
    file_path = "C:/Users/KeikoGolden/KeikoData/resume_corpus.txt"

    text = open(file_path, encoding = 'utf8').read()
    texts = text.split("\n\n\n========DOCUMENT SEPARATOR========\n\n\n")


    texts = random.sample(texts, 100)


    texts = [re.split('\n+',i ) for i in texts]


    texts_short = []
    for text in texts:
        texts_short.extend(text)

    docs = list(nlp.pipe(texts_short))
    
    place_holder = []
    for doc in docs:
        for token in doc:
            if token.similarity(excellent_doc) > similarity_rate:
                if token.pos_ == pos and token.lower_!= modifier:
                    if token.lower_ not in place_holder:
                        place_holder.append(token.lower_)
                        print(token.lower_, token.pos_, token.similarity(excellent_doc))
```
