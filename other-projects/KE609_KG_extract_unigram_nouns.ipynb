```python
#Find all Unigrams
def find_unigram1(df, column_name):

    unigrams = []

    for skill in df[f"{column_name}"]:
        doc = nlp(skill)
        if len(doc)== 1:                               
            unigrams.append(doc)
        else:
            unigrams.append('')
            
    df['unigrams'] = unigrams
    
    return df
```

```python
#Find unigrams except for products and proper nouns.
def find_unigram2(df, column_name):

    unigrams = []

    for skill in df[f"{column_name}"]:
        doc = nlp(skill)
        if len(doc)== 1 and skill== skill.lower():  
            unigrams.append(doc)
        else:
            unigrams.append('')
       
    df['unigrams'] = unigrams
    
    return df
```

```python
#Find unigrams except for products, proper nouns, verbs.
def find_unigram3(df, column_name):

    unigrams = []

    for skill in df[f"{column_name}"]:
        doc = nlp(skill)
        if len(doc)== 1:
            if skill == skill.lower():
                for token in doc:
                    if token.pos_!='VERB': #or token.pos_!="ADJ" or token.pos_ != "ADV":
                        unigrams.append(doc)
                    else: 
                        unigrams.append('')
            else:
                unigrams.append('') 
        else:
            unigrams.append('')

    df['unigrams'] = unigrams
    
    return df
```

```python
#Final. Find unigrams except for products, proper nouns, verbs, adjectives, and adverbs.

def find_unigram4(df, column_name):

    unigrams = []

    for skill in df[f"{column_name}"]:
        doc = nlp(skill)
        if len(doc)== 1: #get unigram
            if skill == skill.lower(): #get non pronouns and non products
                for token in doc:
                    if token.pos_=='VERB': 
                        unigrams.append('')
                    elif token.pos_ == 'ADV':
                        unigrams.append('')
                    elif token.pos_ == 'ADJ':
                        unigrams.append('')  
                    else: 
                        unigrams.append(doc)
            else:
                unigrams.append('') 
        else:
            unigrams.append('')

    df['unigrams'] = unigrams
    
    return df
```

```python
#Final. Find unigrams except for products, proper nouns, verbs, adjectives, and adverbs.

def find_unigram5(df, column_name):

    unigrams = []

    for skill in df[f"{column_name}"]:
        doc = nlp(skill)
        if len(doc)== 1 and skill == skill.lower(): 
            for token in doc:
                if token.pos_=='VERB' or token.pos_ == 'ADV' or token.pos_ == 'ADJ': 
                    unigrams.append('')
                else: 
                    unigrams.append(skill)
        else:
            unigrams.append('') 


    df['unigrams'] = unigrams
    
    return df
```

```python
#Find skill titles with unigram.
import re

def find_titles(df, unigram_column_name, texts):
    
    new_titles = []
    
    for unigram in df[f"{unigram_column_name}"][:50]:
        
        if str(unigram) != str(unigram).upper():
    
            regex = f'({unigram}\s[A-Za-z]+\s[A-Za-z]+)\s'
            matches = re.findall(regex, texts)
            if matches:
                print(unigram)
                print(matches)
                new_titles.append(matches)
            else:
                new_titles.append("not found")
        else:
            new_titles.append("acronym")
    
    df['new_titles'] = new_titles
    
    return df
```
